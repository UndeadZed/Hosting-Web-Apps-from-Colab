```python
%%writefile editor.py
import streamlit as st
import os
from PIL import Image
import torch
import requests
import numpy as np
from models.clipseg import CLIPDensePredT
from PIL import Image
from torchvision import transforms
from matplotlib import pyplot as plt
from PIL import Image, ImageFilter
import requests
import torch
from io import BytesIO
from diffusers import StableDiffusionInpaintPipeline



# load model
model = CLIPDensePredT(version='ViT-B/16', reduce_dim=64)
model.eval();

if 'counter' not in st.session_state:
  st.session_state.counter = 0

# non-strict, because we only stored decoder weights (not CLIP weights)
model.load_state_dict(torch.load('weights/rd64-uni.pth', map_location=torch.device('cpu')), strict=False);



if 'stable_pipeline' not in st.session_state:
    pipeline = StableDiffusionInpaintPipeline.from_pretrained(
    "runwayml/stable-diffusion-inpainting",
    torch_dtype=torch.float16,)
    pipeline = pipeline.to("cuda")
    st.session_state['stable_pipeline'] = pipeline




## making the mask making function

def make_mask(img_path , mask1 ='furniture' , mask2 = 'chandeliers'  , mask3 ='flowers' , mask4 = 'person'):
  # load and normalize image
  input_image = Image.open(img_path)
  transform = transforms.Compose([
      transforms.ToTensor(),
      transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
      transforms.Resize((352, 352)),
  ])
  img = transform(input_image).unsqueeze(0)
  prompts = [mask1, mask2, mask3, mask4]

  with torch.no_grad():
      preds = model(img.repeat(4,1,1,1), prompts)[0]

  # loop over predictions and save each one as image
  for i in range(4):
      # Convert tensor to image
      prediction = torch.sigmoid(preds[i][0]).cpu().numpy()

      # Scale to 0-255
      prediction = (prediction * 255).astype(np.uint8)

      img = Image.fromarray(prediction, 'L')

      # Save the image with prompt as filename
      img.save(f"{prompts[i]}.png")

def turn2mask(path):
  mask_image = Image.open(path)
  mask_image = mask_image.convert("L")
  blurred_mask_image = mask_image.filter(ImageFilter.GaussianBlur(radius=3))
  blurred_mask_image = blurred_mask_image.point(lambda p: 255 if p > 60 else 0)
  return blurred_mask_image

def full_mask(path):
  image = Image.open(path)
  mask = Image.new('L', image.size, 255)
  return mask


if 'prev_img' not in st.session_state:
  st.session_state.prev_img = st.session_state.counter

if st.session_state.prev_img != st.session_state.counter:
  st.session_state.prev_img = st.session_state.counter
  if 'done' in st.session_state:
    st.session_state.done = False

if 'done' not in st.session_state:
  st.session_state.done = True
  if st.session_state.counter == 2:
    make_mask(f"/content/images/0.jpg",mask4 = 'tables')
  elif st.session_state.counter == 0:
    make_mask(f"/content/images/2.jpg",mask4 = 'tables')
  else:
    make_mask(f"/content/images/1.jpg",mask4 = 'tables')

elif st.session_state.done != True:
  st.session_state.done = True
  make_mask(f"/content/images/{st.session_state.counter}.jpg",mask4 = 'tables')

if st.session_state.counter == 2:
    init_image = Image.open(f"/content/images/2.jpg")
elif st.session_state.counter == 0:
    init_image = Image.open(f"/content/images/0.jpg")
else:
    init_image = Image.open(f"/content/images/{st.session_state.counter}.jpg")
init_image = init_image.resize((352, 352))
mask_image = turn2mask(f"/content/furniture.png")
mask_image = mask_image.resize((352, 352))


def edit(prompt):
  prompt = prompt+" , circuar wedding hall tables ,high resolution, realistic"
  image = st.session_state['stable_pipeline'](prompt=prompt, image=init_image, mask_image=mask_image).images[0]
  directory = '/content/weights/'
  image.save(directory + 'output.jpg')

st.title("weddify ✏️")


# Get list of images in folder
folderWithImages = "images"
pathsImages = [os.path.join(folderWithImages,f) for f in os.listdir(folderWithImages)]

# load selected photo it to button
impath = 0
if st.session_state.counter == 0:
  impath = len(pathsImages)-1
else:
  impath = st.session_state.counter-1
photo = pathsImages[impath]

photo = "/content/weights/output.jpg"
if st.session_state.counter == 0:
  original = f"/content/images/{0}.jpg"
elif st.session_state.counter == 2:
  original = f"/content/images/{2}.jpg"
else:
  original = f"/content/images/{1}.jpg"

image  = st.image(original, caption='Original Wedding Hall')
if "messages" not in st.session_state:
    st.session_state.messages = []


if prompt := st.chat_input("say anything"):
    st.session_state.messages.append({"role": "user", "content": prompt})
    with st.chat_message("user"):
        st.markdown(f"Requested changes: {prompt}")
        edit(prompt)

    with st.chat_message("assistant"):
        message_placeholder = st.empty()
        message_placeholder.markdown("Here are the modifications you asked for")
        image  = st.image(photo, caption='Modified Image')
```
